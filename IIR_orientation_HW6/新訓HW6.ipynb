```python
from Bio import SeqIO
input_file = '/content/drive/My Drive/新生訓練/新訓HW6/GRCh38_latest_genomic.fna'
record_dict = SeqIO.index(input_file, "fasta")
S = record_dict["NC_000006.12"].seq[100000:200000]
print(S)
```

```python
 import numpy as np
S = S.lower()
uni_dic = {word:index for index, word in enumerate(sorted(set(list(S))))} # 編碼以利計算markov model
print(uni_dic)
```

```python
# S 從字母轉換至數字(Index)
new_S = []
for i in S:
  new_S.append(uni_dic[i])
print(new_S)
```

# Q1

## Order 0

```python
import math
def order0(arr):
    M = np.zeros(shape=(max(arr) + 1))
    for i in arr:
        M[i] += 1

    T = (M / M.sum(axis=0)) # softmax
    
    return  T # log base 2

print("Order 0: ")
print(uni_dic)
p_list = order0(arr=new_S)
print(p_list)
print("Probability: ", end='')
p = 0
for i in new_S:
  p += math.log(p_list[i], 2)

print(p, "(in log base 2)")
```

## Order 1

```python
import math
def order1(arr):
    # transition matrix
    M = np.zeros(shape=(max(arr) + 1, max(arr) + 1))

    # zip(arr, arr[1:])計算states之間轉換的pair
    for (i, j) in zip(arr, arr[1:]):
        M[i, j] += 1

    T = (M.T / M.sum(axis=1)).T # softmax
    
    return  T # log base 2

print("Order 1: ")
print(uni_dic)
p1_list = order1(arr=new_S)
print(p1_list)
print("Probability: ", end='')
p = math.log(p_list[new_S[0]], 2)
for i in range(1, len(new_S)):
  p += math.log(p1_list[new_S[i-1]][new_S[i]], 2)

print(p, "(in log base 2)")
  
```

## Order 2

```python
import math
def order2(arr):
    # transition matrix
    M = np.zeros(shape=(max(arr) + 1, max(arr) + 1, max(arr) + 1))

    # zip(arr, arr[1:])計算states之間轉換的pair
    relation_count = []
    for i in range(3, len(new_S)):
        relation_count.append([new_S[i-2], new_S[i-1], new_S[i]])
    for (i, j, k) in relation_count:
        M[i, j, k] += 1
    ans = []
    for i in range(len(M)):
      ans.append(list((M[i].T / M[i].sum(axis=1)).T))
    
    return  np.array(ans) # log base 2

print("Order 2: ")
print(uni_dic)
p2_list = order2(arr=new_S)
print(p2_list)
print("Probability: ", end='')
p = math.log(p_list[new_S[0]], 2)
p += math.log(p1_list[new_S[0]][new_S[1]], 2)
for i in range(2, len(new_S)):
  p += math.log(p2_list[new_S[i-2]][new_S[i-1]][new_S[i]], 2)

print(p, "(in log base 2)")
  
```

## Hidden Markov Model

```python
import numpy as np
from hmmlearn import hmm

n_states = 3

model2 = hmm.MultinomialHMM(n_components=n_states, n_iter=100, tol=0.01)
model2.fit(np.atleast_2d(new_S))
print (model2.startprob_)
print (model2.transmat_)
print (model2.emissionprob_)
#print (math.log(math.pow(10, model2.score(np.atleast_2d(new_S))), 2), "(in log base 2)")
print(model2.score(np.atleast_2d(new_S)), "log base on e")
print(model2.score(np.atleast_2d(new_S))/(math.log(2)), "log base on 2")
```
